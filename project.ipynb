 # Mann Whitney Test

```python
from scipy.stats import mannwhitneyu
```

### Attrition vs Distance From Home

```python
a1 = df1["DistanceFromHome"]
a2 = df2["DistanceFromHome"]

stats,p = mannwhitneyu(a1,a2)
```

```python
print(stats,p)
```

```python

#H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition(N)
#H1: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)
# here p<0.5 so,
# We Accept H1 
```

### Attrition vs Income

```python
a1=df1.MonthlyIncome
a2=df2.MonthlyIncome
stat, p=mannwhitneyu(a1,a2)
print(stat, p)

#As the P value is >than 0.05, the H0 is accepted.
#H0: There is no significant differences in the income between attrition (Y) and attirition (N)
#Ha: There is significant differences in the income between attrition (Y) and attirition (N)
```

### Attrition vs  Total Working Years

```python
a1=df1.TotalWorkingYears
a2=df2.TotalWorkingYears
stat, p=mannwhitneyu(a1,a2)
print(stat, p)

#As the P value is again  < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Total Working Years between attrition (Y) and attirition(N)
#Ha: There is significant differences in the Total Working Years between attrition (Y) and attirition (N)
```

### Attrition Vs Years at company

```python
a1=df1.YearsAtCompany
a2=df2.YearsAtCompany
stat, p=mannwhitneyu(a1,a2)
print(stat, p)

#As the P value is again < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Years At Company between attrition (Y) and attirition(N)
#Ha: There is significant differences in the Years At Company between attrition (Y) and attirition (N)
```

### Attrition Vs YearsWithCurrentManager

```python
a1=df1.YearsWithCurrManager
a2=df2.YearsWithCurrManager
stat, p=mannwhitneyu(a1,a2)
print(stat, p)

#As the P value is again  < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Years With Current Manager between attrition (Y) and
#attirition (N)
#Ha: There is significant differences in the Years With Current Manager between attrition (Y) and
#attirition (N)
```

# Seperate T-Test

```python
from scipy.stats import ttest_ind
```

### Attrition Vs Distance From Home

```python
z1=df1.DistanceFromHome
z2=df2.DistanceFromHome
stat, p=ttest_ind(z2,z1)
print(stat, p)

#As the P value is again  < than 0.05, the H0 is  accepted.
#H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition
#(N)
#Ha: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)
```

### Attrition Vs Income

```python
z1=df1.MonthlyIncome
z2=df2.MonthlyIncome
stat, p=ttest_ind(z2, z1)
print(stat, p)

#As the P value is again < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Monthly Income between attrition (Y) and attirition (N)
#Ha: There is significant differences in the Monthly Income between attrition (Y) and attirition (N)
```

### Attrition Vs Yeats At Company

```python
z1=df1.YearsAtCompany
z2=df2.YearsAtCompany
stat, p=ttest_ind(z2, z1)
print(stat, p)

#As the P value is again < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Years At Company between attrition (Y) and attirition(N)
#Ha: There is significant differences in the Years At Company between attrition (Y) and attirition (N)
```

### Attrition Vs Years With Current Manager

```python
z1=df1.YearsWithCurrManager
z2=df2.YearsWithCurrManager
stat, p=ttest_ind(z2, z1)
print(stat, p)

#As the P value is again  < than 0.05, the H0 is rejected and ha is accepted.
#H0: There is no significant differences in the Years With Current Manager between attrition (Y) and
#attirition (N)
#Ha: There is significant differences in the Years With Current Manager between attrition (Y) and
#attirition (N)
```
